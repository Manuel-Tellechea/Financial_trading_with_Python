## ATR Calculation

```python
data['high_low'] = data['High'] - data['Low']
data.tail()
```

```python
data['abs_high_close'] = abs(data['High'] - data['Close'].shift(1))
data.tail()
```

```python
data['abs_high_close'] = abs(data['High'] - data['Low'].shift(1))
data.tail()
```

```python
data['tr'] = data[['high_low', 'abs_high_close', 'abs_low_close']].max(axis=1)
data.tail()
```

```python
data['atr'] = data['tr'].rolling(14).mean()
data = data.drop(columns=["high_low", "abs_high_close", "abs_low_close", "tr"])
data.tail()
```
